### DATA VISUALIZATION

```python
# increase in population size over time
plt.style.use('seaborn')
sns.relplot(data=WP[['Year','Population']], x='Year', y='Population', kind='line', height=10, linewidth=3, color='purple')
plt.title('Population Size')
plt.show()
```

```python
#percentage change in population per yer - measure of population growth
plt.style.use('seaborn')
sns.relplot(data=WP[['Year','ChangePerc']], x='Year', y='ChangePerc', kind='line', height=10, linewidth=3, color='orangered')
plt.title('Population Growth')
plt.show()
```

```python
# population density over time
plt.style.use('fivethirtyeight')
sns.relplot(data=WP[['Year','Density']], x='Year', y='Density', kind='line', height=10, linewidth=3, color='navy')
plt.title('Population Density')
plt.show()
```

```python
url2 = "https://raw.githubusercontent.com/aisha-rm/Kadatemy/main/urRur"
UR = pd.read_csv(url2)
UR.head()
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(10, 10))
plt.plot(UR['year'], UR['rural'],linewidth=3, linestyle="dashed",color='purple', label='Rural Population')
plt.plot(UR['year'], UR['urban'], linewidth=3, linestyle="dashed",color='orangered', label='Urban Population')
plt.title('Rural Population vs Urban Population')

plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.show()
```
